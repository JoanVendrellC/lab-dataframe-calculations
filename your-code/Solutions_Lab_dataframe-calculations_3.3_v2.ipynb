Cleanup Name that contain "Mega".

```python
def clean(x):
    if 'Mega' in x:
        change= x.split('Mega')[1]
        name= 'Mega'+ change
        return name
    else:
        return x
```

```python
Pokemon['Name']=Pokemon['Name'].apply(clean)
```

```python
Pokemon.head(10)
```

Create a new column called A/D Ratio whose value equals to Attack devided by Defense.

```python
Pokemon['A/D Ratio'] = Pokemon['Attack']/Pokemon['Defense']
```

```python
Pokemon.head(10)
```

Identify the pokemon with the highest A/D Ratio

```python
Pokemon[Pokemon['A/D Ratio']==Pokemon['A/D Ratio'].max()]
```

Identify the pokemon with the lowest A/D Ratio.

```python
Pokemon[Pokemon['A/D Ratio']==Pokemon['A/D Ratio'].min()]
```

Create a new column called Combo Type whose value combines Type 1 and Type 2.

```python
Pokemon['Type 1'].isna().sum()
Pokemon['Type 2'].isna().sum()
Pokemon['Type 2']=Pokemon['Type 2'].fillna('')
Pokemon.head(10)
```

```python
def typing(row):
    
    if row['Type 2']=='':
        return row['Type 1']
    else: 
        return row['Type 1']+'-'+row['Type 2']
```

```python
Pokemon['Combo Type'] = Pokemon.apply(typing, axis=1)
Pokemon.head(10)
```

Identify the pokemon whose A/D Ratio are among the top 5.

```python
top5=Pokemon.sort_values(by='A/D Ratio', ascending=False).head(5)
```

For the 5 pokemon printed above, aggregate Combo Type and use a list to store the unique values.

```python
top5_combo_type = []
for value in top5['Combo Type']:
    if value not in top5_combo_type:
        top5_combo_type.append(value)
        
top5_combo_type
```

For each of the Combo Type values obtained from the previous question, calculate the mean scores of all numeric fields across all pokemon.

```python
mean_total = Pokemon.groupby('Combo Type').agg('mean').reset_index()

mean_top5 = mean_total[mean_total['Combo Type'].isin(top5_combo_type)]
mean_top5
```

```python
Orders = pd.read_csv('Orders.csv')
Orders.head(10)
```

```python
Orders.shape
```

```python
Orders = Orders.drop(['Unnamed: 0'],axis=1)
```

How to aggregate the amount_spent for unique customers?

```python
Orders_by_amount = Orders.groupby(['CustomerID']).agg({'amount_spent':sum})
Orders_by_amount
```

```python
Orders_by_amount['amount_spent'].mean()
```

```python
perc95 = np.percentile(Orders_by_amount['amount_spent'],95)
perc75 = np.percentile(Orders_by_amount['amount_spent'],75)
```

```python
def customers_type(x):
    if x > perc95:
        return 'VIP Customer'
    elif x < perc95 and x >= perc75:
        return 'Preferred Customer'
    else:
        return 'Not-preferred Customer'
```

```python
Orders_by_amount['customer_type'] = Orders_by_amount['amount_spent'].apply(customers_type)
```

```python
Orders_by_amount
```

How to select customers whose aggregated amount_spent is in a given quantile range?

```python
country = Orders.groupby(['CustomerID','Country']).agg({'amount_spent':sum})
country
```

```python
country['customer_type'] = country['amount_spent'].apply(customers_type)
country
```

```python
VIPs = country[country['customer_type']=="VIP Customer"]
VIPs
```

```python
VIPs=VIPs.reset_index()
```

```python
VIPs['Country'].value_counts()
```

How to label selected customers as "VIP" or "Preferred"?

```python
VIPs_Preferred = country[(country['customer_type']=="VIP Customer")|(country['customer_type']=="Preferred Customer")]
VIPs_Preferred
```

```python
VIPs_Preferred=VIPs_Preferred.reset_index()
```

```python
VIPs_Preferred['Country'].value_counts()
```
